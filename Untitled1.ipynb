```python
def ECDF(data, x):
    counter = 0
    for v in data:
        if v <= x:
            counter += 1
    return counter / len(data)
```

```python
samples = df['price'] # sepal length
npoints = 500
dx = (samples.max()-samples.min())/npoints

xlist = [samples.min()+dx*i for i in range(npoints)]
ylist = [ECDF(samples, x) for x in xlist]
```

```python
df_ECDF = pd.DataFrame(ylist, columns=['price'],index=xlist)
df_ECDF
```

```python
class SimpleLinReg:

    def __init__(self):
        self.a_ = None
        self.b_ = None

    def fit(self, x_train, y_train):
        assert x_train.ndim == 1, \
            "В данных должен быть один признак"
        assert len(x_train) == len(y_train), \
            "Данные должны иметь одинаковый размер"

        x_mean = np.mean(x_train)
        y_mean = np.mean(y_train)

        self.a_ = (x_train - x_mean).dot(y_train - y_mean) / \
                  (x_train - x_mean).dot(x_train - x_mean)
        self.b_ = y_mean - self.a_ * x_mean

        return self

    def predict(self, x_predict):
        assert x_predict.ndim == 1, \
            "В данных должен быть один признак"
        assert self.a_ is not None and self.b_ is not None, \
            "Модель вначале должна быть обучена"

        return np.array([self._predict(x) for x in x_predict])

    def _predict(self, x_single):
        return self.a_ * x_single + self.b_

    def __repr__(self):
        return "SimpleLinearReg()"
```

```python

# Допустим, у вас есть данные в переменных X и y,
#где X - независимая переменная, y - зависимая переменная

#Построение линейной регрессии
X = df['features/carat'].values.reshape(-1, 1)
y = df['price'].values

# Обучение модели линейной регрессии
lin_reg_model = LinearRegression()
lin_reg_model.fit(X, y)

# Предсказание значений
y_pred_lin_reg = lin_reg_model.predict(X)

# Оценка качества модели по коэффициенту детерминации R^2
r2_lin_reg = r2_score(y, y_pred_lin_reg)
print("R^2 для точной линейной регрессии:", r2_lin_reg)

# Нейронная сеть с одним нейроном
model_nn = Sequential([
Dense(1, input_dim=1)
])

model_nn.compile(optimizer='adam', loss='mean_squared_error')

# Обучение модели нейронной сети
model_nn.fit(X, y, epochs=100, verbose=0)

# Предсказание значений
y_pred_nn = model_nn.predict(X)

# Оценка качества модели по коэффициенту детерминации R^2
r2_nn = r2_score(y, y_pred_nn)
print("R^2 для нейронной сети:", r2_nn)

# Сравнение двух подходов
if r2_lin_reg > r2_nn:
  print("Линейная регрессия точного подхода дает лучший результат.")

else:
  print("Нейронная сеть дает лучший результат.")
```
